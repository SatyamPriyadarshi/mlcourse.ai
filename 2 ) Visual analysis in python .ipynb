```python
# correlation matrix --> used to see wether 2 matrix are correlated to each other or not. 
'''A great method to look at the correlation between 2 or more numerical variables .
first we need to calculate the correlation using .corr() method from sns 
second we plot the correlation matrix (.heatmap() method) using the correlation we calculated earlier.'''
```

```python
df.info()
```

```python
# Syntax
# Drop non-numerical values 
numerical = list(set(df.columns)-set(['Churn',"State",'International plan','Voice mail plan','Area code','Customer service calls']))
# calculating and plotting 
corr_matrix = df[numerical].corr()
sns.heatmap(corr_matrix);
```

```python
# looking at the correlation matrix 
# it can be said that total eve charge and total eve minutes are highly correlated.
# it can be said that total intl charge and total intl minutes are highly correlated.
# it can be said that total day charge and total day minutes are highly correlated.
# it can be said that total night charge and total night minutes are highly correlated.
```

```python
# we need to remove highly correlated values from our numerical features 
numerical = list(set(numerical)-set(['Total eve charge','Total night charge','Total day charge','Total intl charge']))
corr1_matrix= df[numerical].corr()
sns.heatmap(corr1_matrix);
```

```python
# now it can be seen that we have numerical features with us which are not correlated to each other. 
## YAY!!!
```

```python
# t-SNE --> t distributed stochastic neighborhood embedding .

```

```python
# we will delete churn and stste feature as these will not help us to visulaize anything. /
```

```python
# apart from this we will be changing our binary features into numerica features .
```

```python
# Loading libraries 
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
```

```python
 # changing binary features into numerical features 
X = df.drop(['Churn','State'],axis=1)
X["International plan"]= X["International plan"].map({'Yes':1,'No':0})
X['Voice mail plan']= X['Voice mail plan'].map({'Yes':1,'No':0})
```

```python
# normalizing the data using standar scaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# building a tsne representation 
tsne = TSNE(random_state=17)
tsne_repr = tsne.fit_transform(X_scaled)
```

```python
# plotting the tsne representation 
plt.scatter(tsne_repr[:,0],tsne_repr[:,1],alpha=0.5);
```

```python
# Coloring the plot
plt.scatter(tsne_repr[:,0],tsne_repr[:,1],c=df["Churn"].map({False:'orange',True:'blue'}),alpha=0.5);
```

```python
# to better ubderstand the feature let's further color the remqining binary features 
_,axes = plt.subplots(1,2,sharey=True,figsize=(12,8))
for i,name in enumerate(['International plan','Voice mail plan']):
    axes[i].scatter(tsne_repr[:,0],tsne_repr[:,1],c=df[name].map({'Yes':'orange','No':'blue'}),alpha=0.5);
    axes[i].set_title(name);
```

```python
'''Now it is clear that, for example, many dissatisfied customers who canceled their subscription are crowded together in one cluster representing the people with the international plan but no voice mail.

Finally, let's note some disadvantages of t-SNE:

High computational complexity. The implementation in scikit-learn is unlikely to be feasible in a real task. If you have a large number of samples, you should try Multicore-TSNE instead.
The plot can change a great deal depending on the random seed, which complicates interpretation. Here is a good tutorial on t-SNE. In general, you shouldn't make any far-reaching conclusions based on such graphs because it can equate to plain guessing. Of course, some findings in t-SNE pictures can inspire an idea and be confirmed through more thorough research down the line, but that does not happen very often.'''
```
